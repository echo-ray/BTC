```python
%pylab inline
import pandas, pickle
from scipy import stats

tableau20 = pickle.load(open("tableau20.colors","rb"))
rcParams["figure.figsize"] = (12,8)
rcParams["axes.color_cycle"] = tableau20
rcParams["lines.linewidth"] = 2
```

```python
df = pandas.read_csv("./data/data_2014_07_29.csv")
df.time = map(lambda x: float(x[11:-1][:2])*3600.0 + float(x[11:-1][3:5])*60.0 + float(x[11:-1][6:]), df.time)
df["timedelta"] = df.time - df.time.shift(1)
df["timedelta"].iloc[0] = 0.0
```

```python
df.head()
```

```python
mtype = "limit"; outlier = 1.00
timedelta = df.loc[(df.type == mtype) & (df.timedelta < outlier)].timedelta.copy()
timedelta.index = range(len(timedelta))

bins = hist(timedelta, bins=40, normed=True);
ex = expon(scale=timedelta.mean())
plot(bins[1], ex.pdf(bins[1]));
xlim(0.0,outlier);
```

```python
counts, edges = histogram(timedelta)
empirical = concatenate(([0.0],counts.cumsum()))/timedelta.count()
plot(edges, empirical, "--")
plot(edges, ex.cdf(edges));
test_stat = max(abs(empirical - ex.cdf(edges)))
print "kolmogorov-smirnov p-value: %f" % (1.0-stats.ksone.cdf(test_stat, len(edges)))
```
