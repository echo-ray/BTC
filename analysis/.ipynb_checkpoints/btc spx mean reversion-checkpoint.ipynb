```python
%pylab inline
import pandas
import pickle
import statsmodels.tsa.stattools as tsa
import statsmodels.api as sm
import pandas.io.data as web
from datetime import datetime

rcParams["figure.figsize"] = (12,8)
rcParams["axes.grid"] = True
rcParams["axes.color_cycle"] = pickle.load(open("tableau.colors", "rb"))
```

```python
btc = pandas.read_csv("data/btcusddata.csv")
btc = btc[::-1]; btc.index = range(len(btc))

spx = web.DataReader("^GSPC", "yahoo", datetime(2013, 12, 1) , datetime.now())
spx.index = map(lambda idx: str(idx)[:10], spx.index)

btc = btc.loc[btc["Date"].isin(spx.index)]; btc.index = range(len(btc))

btcreturns = (1. + btc["24h Average"].pct_change()).cumprod() - 1.
spxreturns = (1. + spx["Adj Close"].pct_change()).cumprod() - 1.
```

```python
plot(btcreturns)
plot(spxreturns)
```

```python
spxprices = spx["Adj Close"]; spxprices.index = range(len(spxprices))
btcprices = btc["24h Average"]; btcprices.index = range(len(btcprices))

beta = sm.OLS(spxprices, btcprices).fit().params.iloc[0]
spreads = spxprices - beta * btcprices
```

```python
len(spx)
```

```python
tsa.adfuller(spreads)[1]
```

```python
beta
```

```python
print spreads.mean(), spreads.std()
```

```python
def calcspread(spxspot, btcspot):
    spread = spxspot - beta*btcspot
    return spread, spread / spreads.std()
def btccorrect(spxspot):
    return spxspot / beta
```

```python
print calcspread(1990.20, 233.41), btccorrect(1990.20)
```

```python
plot(spreads)
plot(range(len(spreads)), [0.0]*len(spreads), "--", color="r")
plot(range(len(spreads)), [ 1.2 * spreads.std()]*len(spreads), "--", color="r")
plot(range(len(spreads)), [-1.2 * spreads.std()]*len(spreads), "--", color="r")
```
